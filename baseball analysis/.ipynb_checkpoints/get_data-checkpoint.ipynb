```python
import pybaseball 
import matplotlib as plt
import pandas as pd
import seaborn
from pybaseball import statcast
```

```python
pybaseball.cache.enable()
```

```python
from pybaseball import statcast
data_2022 = statcast(start_dt="2022-01-01", end_dt="2022-12-31")
```

```python
print(data_2022.shape)
print(data_2022['player_name'].unique())
print(data_2022['player_name'].nunique())
data_2022.info()
```

```python
data_2023 = statcast(start_dt="2023-01-01", end_dt="2023-12-31")
```

```python
import pickle
import gzip

# save and compress.
with gzip.open('data_2022_zip.pickle', 'wb') as f:
    pickle.dump(data_2022, f)
with gzip.open('data_2023_zip.pickle', 'wb') as f:
    pickle.dump(data_2023, f)
```
